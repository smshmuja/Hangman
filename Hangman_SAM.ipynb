```python
# Dictionary representing the stages of the hangman drawing
hangman = {
  0: '''
    ____________
     |''',
  1: '''
    ____________
     |
     O''',
  2: '''
    ____________
     |
     O
    /''',
  3: '''
    ____________
     |
     O
    / \\''',
  4: '''
    ____________
     |
     O
    / \\
     |''',
  5: '''
    ____________
     |
     O
    / \\
     |
    /''',
  6: '''
    ____________
     |
     O
    / \\
     |
    / \\ '''
}

# List representing the stages of the sunset
sunset_stages = [
  "The sun is shining brightly.",
  "The sun starts to set, the sky is filled with orange hues.",
  "The sky turns red as the sun sinks further.",
  "The sun is almost gone, the sky is a mix of purple and dark blue.",
  "It's twilight, only a sliver of light left.",
  "The sun has set completely, it's a dark night."
]

# List of words for the hangman game
list_of_words = [
  "Algorithm", "Analytics", "Big Data", "Classification", "Data Mining",
  "Eigenvector", "Feature", "Gradient Descent", "Hadoop", "Imputation",
  "Jupyter", "K-Means", "Linear Regression", "Machine Learning",
  "Neural Network", "Outlier", "Python", "Query", "Random Forest",
  "Statistics", "TensorFlow", "Unstructured Data", "Visualization",
  "Web Scraping", "XML", "Yield Prediction", "Z-Score"
]

# Prompt the user for case sensitivity option
option = input("Would you like to play as case sensitive? [y/n]: ")

# If case sensitivity is enabled
if option.lower() == 'y':
  # Select a random word from the list
  word = list_of_words[randrange(len(list_of_words))]
  split_word = list(word)

  # Print the hidden word
  hidden = "_" * len(word)
  hidden_list = list(hidden)
  print(hidden_list)

  # Initialize the number of attempts
  life = 6

  # Loop until the player runs out of attempts or guesses the word
  while life > 0 and hidden_list != split_word:
    print("===========================================================")
    print(sunset_stages[6 - life])
    print(hangman[6 - life])
    print("==> Please enter a case sensitive letter")
    x = input()

    # Check if the guessed letter is in the word
    if word.count(x) > 0:
      # Update the hidden word with the correct guesses
      for i in range(len(word)):
        if x == split_word[i]:
          hidden_list[i] = x

          # Check if the word has been completely guessed
          if hidden_list == split_word:
            print(f"Congratulation, you guessed the word {word}")
            break

      print(f'CORRECT! The word contains the letter {x}, {word.count(x)} times')
    else:
      # Decrease the number of attempts
      life -= 1
      if life > 0:
        print(sunset_stages[6 - life - 1])
        print(hangman[6 - life - 1])
        print(f"WRONG! Number of mistakes left: {life}")
      elif life == 0:
        print("Out of attempts, try again!")
        print("The night has fallen, you couldn't guess the word in time!")
        break

    print(f"Hidden word is {hidden_list}")

# If case sensitivity is disabled
if option.lower() == 'n':
  # Select a random word from the list and convert it to uppercase
  word = list_of_words[randrange(len(list_of_words))]
  word = word.upper()
  split_word = list(word)

  # Print the hidden word
  hidden = "_" * len(word)
  hidden_list = list(hidden)
  print(hidden_list)

  # Initialize the number of attempts
  life = 6

  # Loop until the player runs out of attempts or guesses the word
  while life > 0 and hidden_list != split_word:
    print("===========================================================")
    print(sunset_stages[6 - life])
    print(hangman[6 - life])
    print("==> Please enter a letter")
    x = input().capitalize()

    # Check if the guessed letter is in the word
    if word.count(x) > 0:
      # Update the hidden word with the correct guesses
      for i in range(len(word)):
        if x == split_word[i]:
          hidden_list[i] = x

          # Check if the word has been completely guessed
          if hidden_list == split_word:
            print(f"Congratulation, you guessed the word {word}")
            break

      print(f'CORRECT! The word contains the letter {x}, {word.count(x)} times')
    else:
      # Decrease the number of attempts
      life -= 1
      if life > 0:
        print(sunset_stages[6 - life - 1])
        print(hangman[6 - life - 1])
        print(f"WRONG! Number of mistakes left: {life}")
      elif life == 0:
        print("Out of attempts, try again!")
        print("The night has fallen, you couldn't guess the word in time!")
        break

    print(f"Hidden word is {hidden_list}")
```
